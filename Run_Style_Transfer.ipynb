```python
# Package Import
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
from torch.autograd import Variable
import os
from neural_style_transfer import run_style_transfer
```

```python
#######################################
# Image load
#######################################

# For set up I'll run on cpu, make the code flexible for when we go to gpu
device = torch.cuda.device("cuda" if torch.cuda.is_available() else "cpu")
gpu_available = torch.cuda.is_available()

# desired size of the output image
imsize = 512 if torch.cuda.is_available() else 128  # use small size if no gpu

loader = transforms.Compose([
    transforms.Resize([imsize,imsize]),  # scale imported image
    transforms.ToTensor()])  # transform it into a torch tensor


def image_loader(image_name):
    image = Image.open(image_name)
    # fake batch dimension required to fit network's input dimensions
    image = loader(image).unsqueeze(0)
    if gpu_available:
        image = image.gpu()
    return image    
```

```python
#######################################
# Get Style Images
#######################################
    
# Style images
style_imgs = []
for filename in os.listdir("./monet/"):
    style_imgs.append(image_loader("./monet/"+filename))
```

```python
#######################################
# Image Display
#######################################
#from matplotlib import pyplot

unloader = transforms.ToPILImage()  # reconvert into PIL image

def imshow(tensor):
    image = tensor.cpu().clone()  # we clone the tensor to not do changes on it
    image = image.squeeze(0)      # remove the fake batch dimension
    image = unloader(image)
    return image
    
# Print style image
imshow(style_imgs[0])
```

```python
#######################################
# Pre-trained model and Layer Config
#######################################

# Load pre-trained model
#cnn = models.resnet18(pretrained=True)
cnn_model = models.vgg19(pretrained=True)
cnn = cnn_model.features.eval()

# desired depth layers to compute style/content losses :
content_layers = ['conv_4']
style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5','conv6']
```

```python
#######################################
# Show a couple standard examples
#######################################

## Two different Content Images
# Content Image of a Flower Garden
flower_garden = image_loader("./flower_garden.jpg")

# Content Image of my nieces
my_nieces = image_loader("./twins.jpg")

## Run each
flower_monet = run_style_transfer(cnn, flower_garden, [style_imgs[0]], num_steps=20,
                                 content_layers=content_layers, style_layers=style_layers) # normally we want more epochs

nieces_monet = run_style_transfer(cnn, my_nieces, [style_imgs[0]], num_steps=20,
                                 content_layers=content_layers, style_layers=style_layers)
```

```python
# This style picture worked OK for the flowers because there was similar colors and content
imshow(flower_monet)
```

```python
# For my nieces its not quite as nice, because there are colors which it has trouble representing with the style
imshow(nieces_monet)
```
