```python
# Package Import
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
from torch.autograd import Variable
import os
from neural_style_transfer import run_style_transfer
```

# The plight of a last minute gift giver
Ok so I have twin nieces who are turning 4 this month and I need a good gift....

I wondered if there was something cool I could build them instead of running out and buying another teddy bear... maybe I could make a picture of them but transform it in a cool way. I recently read [this paper](https://www.cv-foundation.org/openaccess/content_cvpr_2016/papers/Gatys_Image_Style_Transfer_CVPR_2016_paper.pdf) on something called neural style transfer, whereby you can recreate the style of one picture with the content of another.

My sister has a few impressionist paintings up in their house, so I figure I can make a picture of my nieces in the style of Monet! I borrowed heavily from this [tutorial](https://pytorch.org/docs/master/tensors.html), and then extended a bit to use multiple Monet paintings and get a little better quality.

Let's get started!

```python
#######################################
# Image load
#######################################

# For set up I'll run on cpu, make the code flexible for when we go to gpu
device = torch.cuda.device("cuda" if torch.cuda.is_available() else "cpu")
gpu_available = torch.cuda.is_available()

# desired size of the output image
imsize = 512 if torch.cuda.is_available() else 128  # use small size if no gpu

loader = transforms.Compose([
    transforms.Resize([imsize,imsize]),  # scale imported image
    transforms.ToTensor()])  # transform it into a torch tensor


def image_loader(image_name):
    image = Image.open(image_name)
    # fake batch dimension required to fit network's input dimensions
    image = loader(image).unsqueeze(0)
    if gpu_available:
        image = image.gpu()
    return image    
```

```python
#######################################
# Get Style Images
#######################################
# Style images
style_imgs = []
for filename in os.listdir("./monet/"):
    style_imgs.append(image_loader("./monet/"+filename))
```

```python
#######################################
# Image Display
#######################################
#from matplotlib import pyplot # Not working in notebooks sometimes?

unloader = transforms.ToPILImage()  # reconvert into PIL image

def imshow(tensor):
    image = tensor.cpu().clone()  # we clone the tensor to not do changes on it
    image = image.squeeze(0)      # remove the fake batch dimension
    image = unloader(image)
    return image
    
# Print style image
imshow(style_imgs[0])
```

```python
#######################################
# Pre-trained model and Layer Config
#######################################

# Load pre-trained model
#cnn = models.resnet18(pretrained=True)
cnn_model = models.vgg19(pretrained=True)
cnn = cnn_model.features.eval()

# desired depth layers to compute style/content losses :
content_layers = ['conv_4']
style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5','conv6']
```

```python
#######################################
# Show a couple standard examples
#######################################

## Two different Content Images
# Content Image of a Flower Garden
flower_garden = image_loader("./flower_garden.jpg")

# Content Image of my nieces
my_nieces = image_loader("./twins.jpg")

# Show Each
imshow(my_nieces)
```

```python
imshow(flower_garden)
```

```python
## Run each
flower_monet = run_style_transfer(cnn, flower_garden, [style_imgs[0]], num_steps=20,
                                 content_layers=content_layers, style_layers=style_layers) # normally we want more epochs

nieces_monet = run_style_transfer(cnn, my_nieces, [style_imgs[0]], num_steps=20,
                                 content_layers=content_layers, style_layers=style_layers)
```

Let's try one last trick to improve quality

## Alternate approach to multiple style images
Intuitively, we really don't need to be limiting the loss from *each* style image, but rather whichever style image fits best. The style image that fits best will be different for different parts of the image. Thus, I'll try an element-wise minimum of differences between the gram matrix of each style image and the output image.

```python
# Keep the greater content layers
content_layers = ['conv_2', 'conv_3', 'conv_4']

# Run output
output = run_style_transfer(cnn, my_nieces, style_imgs, num_steps=10, use_min=True)

# View it
imshow(output)
```

```python
imshow(output)
```

Yowza! Well as you may expect, there were some problems with continuity since we did an element-wise minimum. One (relatively simple) way to correct for this would be to impose a small cost on each pixel which has a different style-image as its minimum from its neighbors. This will act sort of like a regularizer. I do think its interesting that the stlye looks *so different* from the combined style. I'll have to play with this more in the future!

```python
# Save all the output images
import scipy
def save_image(input, filename, basepath="./output_imgs/"):
    image = input.clone().cpu()
    image = image.view(3, imsize, imsize)
    image = unloader(image)
    image.save(basepath+filename+".jpeg")

save_image(nieces_monet, 'nieces')
save_image(flower_monet, 'flowers')
save_image(nieces_monet_multi, 'nieces_multi_style')
save_image(nieces_monet_more_layers, 'nieces_more_layers')
save_image(output, 'nieces_experimental')
```
